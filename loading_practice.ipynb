```python
plt.plot(DA, Arc_slope, 'g^')
plt.ylabel('Average basin slope ArcGIS (degrees)')
plt.xlabel('Watershed area (m^2)')
plt.title('No visible trend between slope and area')
plt.margins(0.2)
plt.show()
```

```python
plt.plot(DA, LSD_slope, 'b^')
plt.ylabel('Average basin slope LSD (radians)')
plt.gca().invert_yaxis()
plt.xlabel('Watershed area (m^2)')
plt.title('No visible trend between slope and area')
plt.margins(0.2)
plt.show()
```

While the titles of the two Slope vs. Area plots say "no visible trend" it is important to note that I do not expect one. The smaller watersheds are not the full watershed (above knickpoint only) and the bigger watershed includes the portions of the tributaries plotted on the same graph. Worth noting: the slope values from arcGIS and LSDtopotools differ - and not systematically.

```python
fig, ax1 = plt.subplots()
t = np.arange(7)
ax1.plot(t, Arc_radians, 'b.')
ax1.set_xlabel('data points')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('ArcGIS derived slope value\n (radians)', color='b')
ax1.tick_params('y', colors='b')
ax1.margins(0.2)

ax2 = ax1.twinx()
ax2.plot(t, LSD_slope, 'g.')
ax2.set_ylabel('LSDTopoTools derived slope value \n (radians)', color='g')
ax2.tick_params('y', colors='g')
ax2.margins(0.2)
plt.annotate('Rio Sabana', xy=(5, -0.1188420571))

plt.title('Variation in slope calculations between the two methods')
plt.show()
```
